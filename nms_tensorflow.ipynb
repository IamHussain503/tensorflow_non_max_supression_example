```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

image1 = plt.imread('/content/drive/MyDrive/test/boundbox/cat.jfif')
image2 = plt.imread('/content/drive/MyDrive/test/boundbox/mh.jpg')


# draw emtpy figure
fig = plt.figure()

# define axis
ax = fig.add_axes([0, 0, 1, 1])

# plot image
plt.imshow(image1)

      
# create rectangular patch
#cat
rect_0 = patches.Rectangle((380, 5), 600, 700, edgecolor='red', facecolor='none', linewidth=2)
rect_1 = patches.Rectangle((300, 100), 300, 410, edgecolor='green', facecolor='none', linewidth=2)
rect_2 = patches.Rectangle((320, 50), 450, 500, edgecolor='blue', facecolor='none', linewidth=2)


    
# add patch
#cat
ax.add_patch(rect_0)
ax.text(188, 376, 'Box0:98%', color='red')
ax.add_patch(rect_1)
ax.text(328, 416, 'Box1:90%', color='green')
ax.add_patch(rect_2)
ax.text(318, 356, 'Box2:82%', color='blue')

# show figure
plt.show()
```
